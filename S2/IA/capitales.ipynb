```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
capitales = pd.read_csv("data/Capitales.csv", sep=";", index_col='Villes')
capitales = capitales.iloc[:, range(0,12)]
capitales
```

```python
from sklearn.decomposition import PCA
model = PCA(svd_solver="full")
capitales_acp = model.fit_transform(capitales)
capitales_acp = pd.DataFrame(capitales_acp)
capitales_acp
```

```python
pd.DataFrame(model.explained_variance_ratio_)
```
