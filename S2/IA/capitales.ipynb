```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
capitales = pd.read_csv("data/Capitales.csv", sep=";", index_col='Villes')
capitales = capitales.iloc[:, range(0,12)]
capitales
```

```python
from sklearn.decomposition import PCA
model = PCA(svd_solver="full")
capitales_acp = model.fit_transform(capitales)
capitales_acp = pd.DataFrame(capitales_acp)
capitales_acp
```

```python
pd.DataFrame(model.explained_variance_ratio_)
```

```python
model.explained_variance_
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
capitales_cah = linkage(capitales, method="ward", metric="euclidean")
plt.title("CAH")
dendrogram(capitales_cah, labels= capitales.index, orientation='right', color_threshold=0)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(distance_threshold=None, n_clusters=5)
capitales_cah = model.fit(capitales)
capitales_cah.labels_
```

```python
CAH = pd.DataFrame({'label' : capitales_cah.labels_}, index=capitales.index)
CAH
```

```python
capitales_acp = capitales_acp.iloc[ : , :2]
model = AgglomerativeClustering(distance_threshold=None, n_clusters=6)
model.fit(capitales_acp)
CAH["label_acp"] = model.labels_
CAH
```

```python
from sklearn.cluster import KMeans
inertie = []
for k in range(1,20) :
    KMean = KMeans(n_clusters=k)
    KMean.fit(capitales)
    inertie.append(KMean.inertia_)
plt.plot(range(1,20), inertie)
```
